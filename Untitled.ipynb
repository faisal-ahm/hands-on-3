```python
import timeit
```

```python
def fun(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

def fun2(n):
    x = 1
    y = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
            y = i + j
    return (x,y)
```

```python
def get_benchmarks():
    benchmarks = [[0], [0]]
    for i in range(1, 1001):
        num = i
        time1 = timeit.timeit(lambda: fun(num), number=1) * 1000000
        time2 = timeit.timeit(lambda: fun2(num), number=1) * 1000000
        benchmarks[0].append(time1)
        benchmarks[1].append(time2)
    return benchmarks
```

```python
benchmarks = get_benchmarks()
benchmark1 = benchmarks[0]
benchmark2 = benchmarks[1]
print(len(benchmark1))
print(len(benchmark2))
```

```python
benchmarks
```

```python
import matplotlib.pyplot as plt

n = list(range(1, 1002))  
plt.figure(figsize=(10, 6))
plt.plot(n, benchmark1, marker='o', linestyle='-', color='b', label='Algorithm 1')
plt.plot(n, benchmark2, marker='o', linestyle='-', color='r', label='Algorithm 2')

plt.xlabel('n', fontsize=14)
plt.ylabel('Time (microseconds)', fontsize=14)

# Add a title
plt.title('Algorithm Benchmark', fontsize=16)

plt.grid(True)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = list(range(1, 1002)) 

bubble_times = benchmarks[0]  
insertion_times = benchmarks[1]  
selection_times = benchmarks[2]  

# Plotting the data
plt.figure(figsize=(10, 6))

plt.plot(sizes, bubble_times, label='Bubble Sort', color='b', marker='o')
plt.plot(sizes, insertion_times, label='Insertion Sort', color='r', marker='o')
plt.plot(sizes, selection_times, label='Selection Sort', color='g', marker='o')

plt.xlabel('Array Size')
plt.ylabel('Time Taken (s)')
plt.title('Sorting Algorithm Performance')
plt.legend()
plt.grid(True)
plt.show()
```
