```python
import timeit
```

```python
def fun(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

x = f(n)
   x = 1;
   y = 1;
   for i = 1:n
        for j = 1:n
             x = x + 1;
        y = i + j;
def fun2(n):
    x = 1
    y = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
            y = i + j
    return (x,y)
```

```python
def get_benchmarks():
    benchmarks = [0]
    for i in range(1, 1001):
        num = i
        time = timeit.timeit(lambda: fun(num), number=1) * 1000000
        benchmarks.append(time)
    return benchmarks
```

```python
benchmarks = get_benchmarks()
```

```python
benchmarks
```

```python
import matplotlib.pyplot as plt

n = list(range(1, 1002))  # Array sizes from 1 to 101
benchmark = benchmarks
plt.figure(figsize=(10, 6))
plt.plot(n, benchmark, marker='o', linestyle='-', color='b', label='Sorting Algorithm')

plt.xlabel('n', fontsize=14)
plt.ylabel('Time (microseconds)', fontsize=14)

# Add a title
plt.title('Algorithm Benchmark', fontsize=16)

plt.grid(True)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
